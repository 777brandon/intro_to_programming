```python
def get_pay_rate() :
    pay_rate = float(input("Please enter the employee's hour pay: "))
    if (pay_rate > 25) :
        print("Hourly pay cannot exceed $25.00. You must enter a number less than or equal to 25.")
        print("Please try again.")
        print()
        pay_rate = get_pay_rate()
    return pay_rate

def get_hours_worked() :
    hours_worked = int(input("Please enter the number of hour the employee worked: "))
    if (hours_worked > 60) :
        print("Hours worked cannot exceed 60. You must enter a number less than or equal to 60.")
        print("Please try again.")
        print()
        hours_worked = get_hours_worked()
    return hours_worked


emp_number = input("Please enter the employee number: ")
hourly_pay = get_pay_rate()
num_hours = get_hours_worked()
if num_hours > 35 :
    overtime_hours = num_hours - 35
    overtime_pay = hourly_pay * 1.5 * overtime_hours
    regular_hours = 35
else :
    overtime_hours = 0
    overtime_pay = 0
    regular_hours = num_hours
    
regular_pay = regular_hours * hourly_pay
weekly_pay = regular_pay + overtime_pay

print(f"Employee number:\t{emp_number}\nHourly pay:\t\t${hourly_pay:.2f}")
print(f"Hours worked:\t\t{num_hours}\nWeeks pay:\t\t${weekly_pay:.2f}")
```
